```python
import pandas as pd
```

```python
df=pd.read_excel(r'c:\Users\ASUS\Downloads\Store_Sales_Dataset.xlsx.xlsx')
```

```python
#1 View first few rows
print(df.head())
```

```python
# Basic info
df.info()
```

```python
#2 Basic Exploration
# Total number of records
print("Total Records:", len(df))
```

```python
# Total quantity sold
print("Total Quantity Sold:", df["Quantity"].sum())
```

```python
# Total revenue
print("Total Revenue (₹):", df["Total Amount (₹)"].sum())
```

```python
# Average sale value
print("Average Sale (₹):", df["Total Amount (₹)"].mean())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
#3 Filtering
# Sales by a particular salesperson
print(df[df["Salesperson"] == "A. Singh"])
```

```python
# All Electronics category sales
print(df[df["Product Category"] == "Electronics"])
```

```python
# Sales above ₹2000
print(df[df["Total Amount (₹)"] > 2000])
```

```python
#4 Grouping and Aggregation
# Total sales by product category
print(df.groupby("Product Category")["Total Amount (₹)"].sum())
```

```python
# Average sales per store
print(df.groupby("Store Location")["Total Amount (₹)"].mean())
```

```python
# Total quantity sold by salesperson
print(df.groupby("Salesperson")["Quantity"].sum())
```

```python
# Count of sales by payment method
print(df["Payment Method"].value_counts())
```

```python
# Total revenue per customer
print(df.groupby("Customer Name")["Total Amount (₹)"].sum())
```

```python
#5 Sorting Data
# Sort by total amount (descending)
print(df.sort_values("Total Amount (₹)", ascending=False))
```

```python
# Sort by location and date
print(df.sort_values(["Store Location", "Date"]))
```

```python
#6: Date-Based Analysis
# Total sales per day
daily_sales = df.groupby("Date")["Total Amount (₹)"].sum()
print(daily_sales)
```

```python
# Day with highest revenue
print("Highest revenue day:", daily_sales.idxmax(), "with ₹", daily_sales.max())
```

```python
# Total sales per week
df["Date"] = pd.to_datetime(df["Date"])
df["Week"] = df["Date"].dt.isocalendar().week
print(df.groupby("Week")["Total Amount (₹)"].sum())
```

```python
#7: Top Performers
# Top 3 selling products
print(df.groupby("Product Name")["Total Amount (₹)"].sum().sort_values(ascending=False).head(3))
```

```python
# Top salesperson
print(df.groupby("Salesperson")["Total Amount (₹)"].sum().sort_values(ascending=False).head(1))
```

```python

# City with most revenue
print(df.groupby("Store Location")["Total Amount (₹)"].sum().sort_values(ascending=False).head(1))
```

```python
# 9: Visualization
import matplotlib.pyplot as plt
```

```python
# Bar chart - total sales by category
df.groupby("Product Category")["Total Amount (₹)"].sum().plot(kind="bar", title="Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales (₹)")
plt.show()
```

```python
# Pie chart - sales by payment method
df["Payment Method"].value_counts().plot(kind="pie", autopct='%1.1f%%', title="Payment Methods")
plt.show()
```

```python
# Line chart - daily total revenue
daily_sales.plot(kind="line", marker="o", title="Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Revenue (₹)")
plt.show()
```
